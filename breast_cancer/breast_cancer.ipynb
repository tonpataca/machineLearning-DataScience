```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data = load_breast_cancer(as_frame=True)
cancer_df = cancer_data.data
cancer_df['target'] = cancer_data.target
```

```python
cancer_df.head()
```

```python
cancer_df.shape
```

# number of missing values

```python
cancer_df.isnull().count().sum()
```

```python
benign_count = (cancer_df['target'] == 0).sum()
malignant_count = (cancer_df['target'] == 1).sum()
```

## Calcular e mostrar o numero cancer malignos e benignos

```python
print(f'Cancer Benignos: {benign_count}')
print(f'Cancer malignos: {malignant_count}')
```
